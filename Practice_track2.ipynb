```python
import numpy as np
import grid2op
from grid2op.Agent import DoNothingAgent

class DummyAgent(DoNothingAgent):
    def __init__(self, action_space):
        super().__init__(action_space)

multimix_env = env
my_agent = DummyAgent(multimix_env.action_space)
NUM_EPISODES = 1
MAX_STEPS = 2016

# Shuffle scenarios in all mixes
for mix in multimix_env:
    # Using numpy to shuffle list
    def shuff(x):
        lx = len(x)
        s = np.random.choice(lx, size=lx, replace=False)
        return x[s]

    # Call 'regular' environment scenario shuffle
    mix.chronics_handler.shuffle(shuffler=shuff)

# Enter training loop
for i in range(NUM_EPISODES):
    # Pick a random mix on reset
    multimix_env.reset(random=True)
    
    # Print some info on current episode
    mix_name = multimix_env.name
    chronic_name = multimix_env.chronics_handler.get_name()
    print ("Episode [{}] - Mix [{}] - Chronic [{}]".format(i, mix_name, chronic_name))
    
    done = False
    obs = multimix_env.current_obs
    reward = 0.0
    step = 0
    while done is False and step < MAX_STEPS:
        agent_action = my_agent.act(obs, reward, done)
        obs, reward, done , info = multimix_env.step(agent_action)  
        
        #########################################################
        prod_p = obs.prod_p
        load_p = obs.load_p
        prod_v = obs.prod_v
        prod_v_forcasted = obs.get_forecasted_inj(time_step=1)[1]
        load_v = obs.load_v
        load_q = obs.load_q
        prod_p_forcasted = obs.get_forecasted_inj(time_step=1)[0]
        load_p_forcasted = obs.get_forecasted_inj(time_step=1)[2]
        check_1 = prod_p - prod_p_forcasted
        check_2 = load_p - load_p_forcasted
        #########################################################
        
        #print(obs.timestep_overflow)
        #print(np.round(obs.a_or - obs.a_ex))
        #print(obs.rho)
        print(obs.topo_vect)
        print(obs.load_p)
        #print(obs.time_next_maintenance)
        #print(load_v-load_v_forcasted)
        #print(np.sum(load_q))
        #print(np.sum(prod_p_forcasted)-np.sum(load_p_forcasted))
        #print((np.absolute(obs.p_or - obs.p_ex)))
        #print(np.sum(load_p_forcasted))
        #print(np.sum(np.absolute(check_1)))
        #print(np.sum(np.absolute(check_2)))
        #print(obs.time_before_cooldown_line)
        #print(obs.time_before_cooldown_sub)
        #print(obs.simulate(agent_action,time_step=1)[3]['exception'])
        step += 1
    print(step)
```
